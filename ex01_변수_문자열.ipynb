#### 문자열의 인덱싱
- 인덱싱(indexing) : 하나하나를 가리킬 때 사용 (기호 : 대괄호 [ ])
- 인덱스(index) : 색인, 지표, 지수와 같은 뜻으로, 순서를 나타내는 영어단어

인덱스 번호는 앞에서 0부터 시작한다

```python
# 대문자 S를 출력해보자
s = 'My name is MS'
print(s[12])
# 앞에서부터 세어주는게 아니고 뒤에서부터 세어서 찾아보자
print(s[-1])
# 뒤에서부터 셀 때는 순서가 -1에서 시작한다.
```

#### 문자열의 슬라이싱
- 슬라이싱(slicing) : 범위를 지정하여 어디부터 어디까지 잘라냄
- 기호는 대괄호 ([ ])를 사용, 시작점과 끝점을 구분하는 기호로 콜론(:)을 사용

```python
# 'My', 'name', 'MS' 라는 문자열 가져오기
# s[시작 인덱스 : 끝 인덱스 + 1]

s = 'My name is MS'
print(s[0:2])
print(s[:2])
print(s[3:7])
print(s[11:13])
print(s[-2:])  
```

```python
# is MS 가져오기

s = 'My name is MS'
print(s[8:])  # = s[8:13]   끝까지 가져올거면 끝 점은 생략해도 된다
```

```python
# My name 가져오기

s = 'My name is MS'
s[:7]         # = s[0:7]   처음부터 가져올거면 시작 점을 생략해도 된다
```

```python
# 모든 문자열 가져오기
s = 'My name is MS'
print(s[0:13])
print(s[:])
```

```python
print('제 이름은 민수입니다를 영어로 쓰면 >> ',s)
```

#### 문자열의 특정 부분 초기화(formatting)

```python
# 1. format 함수 사용
# 문장 안에 초기화 될 부분에 중괄호{}를 넣어줌
# 문자열 기호 끝에 .format()을 붙이고 ()안에 연결할 값을 넣어주기
```

```python
month = 2
day = 15
s = '오늘은 {}월 {}일 입니다.'.format(month,day)
print(s)
```

```python
# 2. f문자열 포매팅 사용  -> 최신식
# 문자열 기호 맨 앞에 접두사로 f를 붙임
# 문장 안에 초기화 부분은 {}로 연결하고 {} 안에 변수 값 넣기
```

```python
month = 2
day = 15
s = f'오늘은 {month}월 {day}일 입니다.'
print(s)
```

- strip(): 문자열 양 옆에 있는 공백 제거

```python
s = '   python is very easy   '
s
```

```python
s.strip() # 양 옆 공백 제거
```

```python
s.rstrip() # 오른쪽 공백만 제거
```

```python
s.lstrip() # 왼쪽 공백만 제거
```
