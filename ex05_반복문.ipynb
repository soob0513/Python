#### while문
조건식이 True일 경우 실행문장 반복

```python
number = 1
while number <= 3 :
    print(number)
    number += 1
```

- while문 예제 : while문을 사용해서 '파이썬 최고!'를 13번 출력해보자.

```python
num = 0   # while문의 반복횟수를 제한할 수 있는 변수
while num < 13 :
    num += 1
    print(f'{num} 파이썬 최고!')
```

1. while문 실습예제) 두 개의 정수를 입력 받아서 더하는 코드를 작성하시오.(단, 두 개의 정수가 모두 0이 들어오기 전까지 반복)

```python
# 힌트!
# 1. 정수 두 개 입력 먼저 받아보기
# 2. 결과 화면을 살펴보니, 입력하는 것도 반복이 되고 있네요? --> 어떻게 해야할까
# 3. 반복문 안에서 입력값에 0, 0이 들어왔는지 '판단'  --> break로 반복문 빠져나가기
# 4. 0, 0이 아니라면 두 정수의 합을 계산하는 코드 작성
```

```python
while True :
    num1 = int(input('첫 번째 정수 입력 >> '))
    num2 = int(input('두 번째 정수 입력 >> '))
    if num1 == num2 == 0 :
        print('프로그램이 종료되었습니다.')
        break
    else : 
        print(f'두 정수의 합 : {num1+num2}')
```

```python
# 선생님풀이
while True : 
    num1 = int(input('정수입력1 >>'))
    num2 = int(input('정수입력2 >>'))
    if num1 == num2 == 0:
        print('프로그램이 종료되었습니다.')
        break
    result = num1 + num2
    print('두 정수의 합 : ',result)   # else는 꼭 안 써도 된다!
```

2.  while문 실습 예제) 다이어트 관리 프로그램을 만들 예정입니다. 현재 몸무게와 목표 몸무게를 입력받고 주차별 감량 몸무게를 입력받으세요. 목표 몸무게를 달성하면 축하한다는 문구를 출력하고 입력을 멈추세요!

```python
a = int(input('현재 몸무게 : '))
b = int(input('목표 몸무게 : '))
d = 0
while True : 
    d+=1
    w = int(input(f'{d}주차 감량 몸무게 : '))
    a = a-w
    if a <= b :
        print(f'{a}kg 달성!! 축하합니다!')
        break
```

```python
# 힌트!
# 1. 현재 몸무게와 목표 몸무게를 입력받는다!
# 2. 반복해서 수행할건 감량할 몸무게니까 몇주에 걸쳐 감량이 되는지 세어주는 변수 필요
# 3. 현재 몸무게가 목표 몸무게보다 작아지면 다이어트에 성공한거니까 반복문 out!
```

```python
# 선생님풀이
now = int(input('현재 몸무게 >> '))
goal = int(input('목표 몸무게 >> '))
cnt = 0
while True : 
    cnt += 1
    week = int(input(f'{cnt}주차 감량 몸무게 : '))
    now = now - week
    print(f'{cnt}주차 감량 후 현재 몸무게는 {now}kg입니다.')
    if now <= goal :
        print(f'{now}kg 달성!! 축하합니다!! 다이어트 성공!')
        break
```

1. for문 실습 : 5명에 대한 정보처리기사 자격증 시험 점수 리스트가 있습니다. 이 때, 각 점수가 합격인지 불합격인지 판별하여 출력하시오. (60점 이상 합격)


```python
score_list = [90, 45, 70, 60, 55]
s = 1
for score in score_list : 
    if score >= 60 :
        print(f'{s}번 학생은 합격입니다.')
    else : 
        print(f'{s}번 학생은 불합격입니다.')
    s += 1
```

```python
score_list = [90, 45, 70, 60, 55]
s = 0
for score in score_list : 
    s += 1
    if score >= 60 :
        print(f'{s}번 학생은 합격입니다.')
    else : 
        print(f'{s}번 학생은 불합격입니다.')
```
