### 연산자의 종류
1. 산술 연산자 : +, -, *, /, //, %
2. 지수 연산자 : **
3. 대입(복합)연산자 : =, +=, -=, *=, /=, //=, %=
4. 관계(비교)연산자 : >, <, >=, <=, ==, !=  --> 비교를 통한 결과가 참 혹은 거짓으로 
5. 논리 연산자 : and, or, not
6. 삼항 연산자 : a if 조건식 else b
7. 멤버 연산자 : in, not in

#### 치환! : 자리를 바꾸다.
- A와 B는 각각 볼펜을 들고 있다. A는 빨간색 펜, B는 파란색 펜을 들고 있을 때, 들고 있는 손만을 이용해서 A는 파란색, B는 빨간색 펜을 가질 수 있도록 하려면 어떻게 해야할까?

```python
a = 3
b = 7

# 치환방법 1. 
# a에는 b 데이터(7)이, b에는 a 데이터(3)이 담기도록

temp = a
a = b
b = temp
print(a)
print(b)
```

```python
# 치환방법 2.
# 파이썬은 쉽고 간편해요!

a = 3
b = 7

print(f'a:{a}')
print(f'b:{b}')

a,b = b,a
print(f'a:{a}')
print(f'b:{b}')
```

### 논리 연산자
- True, False의 논리값(boolean값)을 반환하는 연산자

- not 연산자

```python
a = 3
b = 7
print(f'a < b : {a<b}')
print(f'not a < b : {not a<b}')   # not : 부정연산자 -> True or False의 반대 값 출력
```

- and 연산자 : 두 조건이 모두 참(True)일 경우에만 참을 반환

```python
print(3 > 5)  # --> False
print(10 == 20)  # -->False

# 둘 다 False

print(3 > 5 and 10 == 20)
```

```python
print(3 < 5)  # --> True
print(10 == 20)  # -->False

# 하나만 False

print(3 < 5 and 10 == 20)
```

```python
print(3 < 5)  # --> True
print(10 != 20)  # --> True

# 둘 다 True

print(3 < 5 and 10 != 20)
```

- or 연산자 : 두 조건 중 하나라도 참(True)이면 참 반환

```python
print(3 < 5)  # --> True
print(10 == 20)  # -->False

# 하나만 False

print(3 < 5 or 10 == 20)
```

```python
print(3 > 5)  # --> False
print(10 == 20)  # -->False

# 둘 다 False

print(3 > 5 or 10 == 20)
```

### 삼항연산자
- a if 조건식 else b
- (True가 나왔을 때 실행할 문장) if 조건식 --> True or False else (False가 나왔을 때 실행할 문장)

```python
score = 80
print('합격입니다.') if score>=80 else print('불합격입니다.')
```

```python
score = 30
print('합격입니다.') if score>=80 else print('불합격입니다.')
```
