2. 대박이는 마스크를 대량 구매했습니다. 주변 친구들에게 소분해서 나눠주기 위해 포장지를 사려 합니다. 포장지는 마스크를 각각 8개, 5개씩 넣을 수 있는 8개들이, 5개들이 2종류가 있습니다. 8개들이는 포장지가 비싸서 반드시 꽉꽉 채워 포장할 때만 사용하고, 만약 마스크가 남는다면 5개들이에 포장하려고 합니다. 과연 포장지는 각각 몇 개가 필요할까요? (단, 마스크는 남는 마스크 없이 전부 포장해야 합니다.


```python
# 수빈
mask = int(input("마스크 개수 입력>> "))
num8 = mask//8
num5 = mask%8
if num5 == 0 :
    print(f"필요한 8개들이 개수: {num8}, 5개들이 개수: 0 입니다.")
else :
    if num5<=5 :
        print(f"필요한 8개들이 개수: {num8}, 5개들이 개수: 1 입니다.")
    else : 
        print(f"필요한 8개들이 개수: {num8}, 5개들이 개수: 2 입니다.") 
```

```python
#힌트!
# 1. 마스크의 개수를 사용자에게 입력받는다.
# 2. 8개들이, 5개들이, 나머지 마스크가 들어갈 변수를 초기화해준다.
# 3. 마스크의 개수가 8개 이상일 때는 8개들이와 5개들이 포장지가 몇 개 필요한지 구하고
# 4. 남은 마스크의 개수가 2개, 3개 처럼 애매하게 남았다면 5개들이 포장지 + 1
# 5. 남은 마스크의 개수가 8개 이하일 때는 8개들이는 필요 없겠죠?
# 6. 마지막에 출력문으로 8개들이와 5개들이 포장지 개수를 입력하면 끝!
```

```python
mask = int(input("마스크 개수 입력>> "))
a = mask//8
if mask%8 <=5 :
    if mask % 8 ==0 :
        b = 0
    else : 
        b=1
else : 
    b=2
print(f'필요한 8개들이 개수: {a}, 5개들이 개수 : {b}입니다.')
```

```python
mask = int(input("마스크 개수 입력>> "))
num8 = mask//8
num5 = mask%8
if num5 == 0 :
    b = 0
else :
    if num5<=5 :
        b = 1
    else : 
        b = 2
print(f'필요한 8개들이 개수: {num8}, 5개들이 개수 : {b}입니다.')
```

```python
# 선생님풀이
countOfMask = int(input('마스크의 개수 >> '))

ei = fi = re = 0
if countOfMask >=8 :   # 마스크가 8개 이상일 때
    ei = countOfMask // 8   # 8개들이 포장지 개수
    fi = countOfMask % 8 // 5   # 5개들이 포장지 개수
    re = countOfMask % 8 % 5   # 전부 포장하고 남은 마스크 개수
    if re != 0 :   # 남은 마스크의 개수가 몇 개든 상관없이 5개들이 포장지 하나 추가!
        fi += 1
else : #마스크가 8개 미만일 때
    fi = countOfMask // 5
    re = countOfMask % 5
    if re != 0 :
        fi += 1

print(f"필요한 8개들이 개수 : {ei}, 5개들이 개수 : {fi} 입니다.")
```

2. 변수 score에 점수를 입력 받아서 학점 관리 프로그램을 만들어보자.

```python
score = int(input('점수 입력>> '))
if score >=90 : 
    print(f'{score}점은 A학점 입니다.')
elif score >= 80 :
    print(f'{score}점은 B학점 입니다.')
elif score >= 70 : 
    print(f'{score}점은 C학점 입니다.')
elif score >= 60 :
    print(f'{score}점은 D학점 입니다.')
else :
    print(f'{score}점은 F학점 입니다.')
```

```python
score = int(input('점수 입력>> '))
if score >=90 : 
    a = 'A'
elif score >= 80 :
    a = 'B'
elif score >= 70 : 
    a = 'C'
elif score >= 60 :
    a = 'D'
else :
    a = 'F'
    
print(f'{score}점은 {a}학점 입니다.')
```

```python
print('''
================================================
================---- 자판기 ----================
================================================
== 1.콜라 == 2.사이다 ==  3.물  =======insert===
==--------==----------==--------=======------===
==   600  ==    800   ==  1000  ======= O00원===
=======================================------===
================================================
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==+++++++++++++++     SMHRD    +++++++++++++++==
==+++++++++++++++              +++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
==++++++++++++++++++++++++++++++++++++++++++++==
================================================
================================================
  ===                                      ===
  ===                                      ===
''')
ic = int(input("insert coin >> "))
menu = int(input("메뉴 선택 >>"))
if menu == 1 :
    if ic < 600 : 
        print('잔액이 부족합니다.')
    else : 
        print(f'잔돈 >> 1000원 {(ic-600)//1000}개, 500원 {(ic-600)%1000//500}개, 100원 {(ic-600)%1000%500//100}개')
elif menu == 2 : 
    if ic <800 :
        print('잔액이 부족합니다.')
    else : 
        print(f'잔돈 >> 1000원 {(ic-800)//1000}개, 500원 {(ic-800)%1000//500}개, 100원 {(ic-800)%1000%500//100}개')    
elif menu == 3 :
    if ic < 1000 :
        print('잔액이 부족합니다.')
    else : 
        print(f'잔돈 >> 1000원 {(ic-1000)//1000}개, 500원 {(ic-1000)%1000//500}개, 100원 {(ic-1000)%1000%500//100}개')
else : 
    print('메뉴를 잘못 선택하셨습니다. ')   
```

```python
# 힌트!
# 1. 돈과 메뉴를 입력 받는다.
# 2. 집어넣은 돈과 상관 없이 메뉴 가격에 따른 거스름돈을 구한다
# 3. 거스름돈이 양수인지, 음수인지에 따른 조건문을 만들고
# 4. 음수일 경우에는 잔액이 부족합니다.
# 5. 양수일 경우에는 1000원짜리, 500원짜리, 100원짜리 의 개수를 구해준다.
# 6. 마지막에 거스름돈을 출력해주면 끝!
```

```python
# 선생님 풀이
money = int(input('insert coin >> '))
menu = int(input('메뉴를 선택하세요 >> '))

# co = 600
# si = 800
# wa = 1000
# change -> 거스름돈
change = 0   # 자판기를 새로 이용할 때마다 거스름돈이 초기화 될 수 있도록
if menu == 1 :  # 콜라
    change = money - 600
elif menu == 2 :  # 사이다
    change = money - 800
elif menu == 3 :   # 물
    change = money - 1000
else : 
    print('메뉴를 잘못 선택하셨습니다. ')
    
#ch = 천원, o는 오백원, b = 백원
ch = o = b = 0

if change < 0 : 
    print('잔액이 부족합니다.')
else : 
    ch = change // 1000
    o = change % 1000 // 500
    b = change % 1000 % 500 //100
    print(f'잔돈 >> 1000원 {ch}개, 500원 {o}개, 100원 {b}개')
```
